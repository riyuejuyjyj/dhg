```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# 定义数据
X = np.random.randn(1000,3)
print(X.shape)
```

```python
# 定义PCA对象，n_com表示最后保留几维
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 画图
fig = plt.figure(figsize=(12,4))
# 添加子图121是图的位置
ax1 = fig.add_subplot(121,projection="3d")
ax1.scatter(X[:,0],X[:,1],X[:,2],c="g") # type: ignore
ax1.set_title("Before PCA(3d)")
ax1.set_xlabel("Feature1")
ax1.set_ylabel("Feature2")
ax1.set_zlabel("Feature3")


ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1],c="g") # type: ignore
ax2.set_title("After PCA(2d)")
ax2.set_xlabel("pc1")
ax2.set_ylabel("pc2")

plt.show()
```

```python
#定义本质特征
n = 1000
pc1 = np.random.normal(0,1,n)
pc2 = np.random.normal(0,0.2,n)
noise = np.random.normal(0,0.1,n)

# 手动构造特征数据(基于主成分做叠加组合)
x1 = pc1 - pc2
x2 = pc1 + pc2
x3 = pc2 + noise

X = np.vstack((x1,x2,x3)).T
print(X.shape)
```

```python
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 画图
fig = plt.figure(figsize=(12,4))
# 添加子图121是图的位置
ax1 = fig.add_subplot(121,projection="3d")
ax1.scatter(X[:,0],X[:,1],X[:,2],c="g") # type: ignore
ax1.set_title("Before PCA(3d)")
ax1.set_xlabel("Feature1")
ax1.set_ylabel("Feature2")
ax1.set_zlabel("Feature3")


ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1],c="g") # type: ignore
ax2.set_title("After PCA(2d)")
ax2.set_xlabel("pc1")
ax2.set_ylabel("pc2")

plt.show()
```
