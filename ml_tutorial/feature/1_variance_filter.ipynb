```python
import numpy as np
```

```python
# 生成特征
x1 = np.random.randn(100)
print(x1.var())
```

```python
x2 = np.random.normal(5,0.1,100)
print(x2.var())
```

```python
X = np.vstack((x1,x2)).T
print(X)
```

```python
from sklearn.feature_selection import VarianceThreshold

vt = VarianceThreshold(threshold=0.01)
```

```python
X_filter=vt.fit_transform(X)
print(X_filter.shape)
```
